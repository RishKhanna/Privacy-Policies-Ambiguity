```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("PolicyCats.csv")
```

```python
val= pd.DataFrame(data.Category.value_counts())
val.index
```

```python
plt.figure(figsize=(20,12))
           
plt.pie(val.Category, labels=val.index)
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(20,12))
           
plt.bar(val.index, val.Category, label = 'Number of Documents')
plt.legend(loc='upper right')
```
